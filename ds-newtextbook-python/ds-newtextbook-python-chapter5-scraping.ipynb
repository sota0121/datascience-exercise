```python
# =====================================
# 5.1.2. スクレイピング環境の準備
# =====================================
# requests, beautifulsoup4 を利用する
import requests
from bs4 import BeautifulSoup as bs
```

```python
# =====================================
# 5.1.3. Webページをダウンロード
# =====================================
# requestsによりWebサーバーから情報を取得
url_codezine = 'https://codezine.jp/'
res = requests.get(url_codezine)
print(type(res))
print(res.status_code)
```

```python
html_txt = res.text
for line in html_txt.split('\n'):
    print(line)
```

```python
# =====================================
# 5.1.4. Webページから要素を抜き出す
# =====================================
# BeautifulSoupによりHTMLの要素を解析（基本）
# 入力テキストとパーサー(ここではhtml)を指定して、解析済みオブジェクトを生成
soup = bs(html_txt, 'html.parser')
print('** <title>タグの情報 : soup.title\n', soup.title, '\n')
print('** <h1>タグの情報 : soup.h1\n', soup.h1, '\n')
print('** h1 タグの中の a タグの中の img タグの中の alt 属性 : soup.h1.a.img[\'alt\']\n', soup.h1.a.img['alt'], '\n')
```

```python
# BeautifulSoupによりHTMLの要素を解析（その他いろいろ）
# ページ内の全 aタグを取得して、情報を表示する
atags = soup.find_all('a')
print('aタグの総数 : ', len(atags), '\n')
print('--- aタグ最初の５個を表示する')
for i, atag in enumerate(atags[:5]):
    print('element [{0}]'.format(i))
    print('タイトル : ', atag.text)
    print('リンク   : ', atag['href'])
    print()
```

```python
# =====================================
# 5.1.5. 記事の一覧を抜き出す
# > https://codezine.jp/article/tag/223
# > は記事の一覧である。ここから各記事の
# > ・日付：記事の公開日付
# > ・タイトル：記事のタイトル
# > ・URL：記事の本文のURL
# > ・タグ：記事のタグ情報
# =====================================
# まずはHTMLの構造把握
url_article_list = 'https://codezine.jp/article/tag/223'
res = requests.get(url_article_list)
with open('codezine_articles.html', 'w') as f:
    f.write(res.text)
# まとめ
# > 記事全体が <ul class="catList"> の中にある (l348付近)
# > 1つの記事はその中の <li> 単位
# > <div class="day"> の中に日付
# > <h2> の中の <a> の中にタイトルとリンク先
# > <ul class="tag"> の中にタグ情報（タグ名とリンク）が <li> 単位で存在する
```

```python
from datetime import datetime
# parse html
soup = bs(res.text, 'html.parser')
# CSSセレクタで <ul class="catList"><li> を取得
lis = soup.select('ul.catList > li')
articles = []
for li in lis:
    # 記事公開日付取得
    published_str = li.find('div', class_='day').text.strip()
    published = datetime.strptime(published_str, '%Y/%m/%d')
    # 記事タイトル取得
    h2_tag = li.find('h2')
    article_title = h2_tag.text
    # 記事本文のURL取得
    article_url = h2_tag.a['href']
    # 記事のタグリストを取得
    tag_list = li.select('ul.tag > li')
    tags = [tag.text.strip() for tag in tag_list]
    # 出力情報を辞書としてまとめる
    article = {
        'published': published,
        'title': article_title,
        'url': article_url,
        'tags': tags
    }
    articles.append(article)
# 解析結果出力
articles[:3]
```

```python
import pandas as pd
df = pd.DataFrame(articles)
df
```
