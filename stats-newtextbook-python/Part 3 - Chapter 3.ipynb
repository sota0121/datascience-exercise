# matplotlib, seabornによるデータの可視化

```python
# prepare
import numpy as np
import pandas as pd
%precision 3
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set()
```

```python
x = np.arange(-10, 10, 0.01)
y = np.sin(x)
plt.plot(x, y, color='black')
plt.title('lineplot seaborn')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# seabornによるヒストグラム
fish_data = np.array([2, 3, 3, 4, 4, 4, 4, 5, 5, 6])
fish_data
```

```python
sns.distplot(fish_data, bins=5, color='red', kde=False)
```

```python
# seabornによるヒストグラム（kde : カーネル密度推定による平滑化）
sns.distplot(fish_data, bins=5, color='red')
```

```python
# 2変量データに対するヒストグラム
fish_data = pd.read_csv('./samples/3-3-2-fish_multi_2.csv')
fish_data
```

```python
len_a = fish_data.query('species == "A"')['length']
len_b = fish_data.query('species == "B"')['length']
sns.distplot(len_a, bins=5, color='orange')
sns.distplot(len_b, bins=5, color='red')
```

```python
# ========================================
# Tips : 多変量データを図示するコードの書き方
# ========================================
# sns.関数名(
#     x = 'x軸の列名',
#     y = 'y軸の列名',
#     data = データフレーム,
#     その他引数
# )
```

```python
# seaborn 可視化関数 一覧
# - 箱ひげ図: boxplot
# - バイオリンプロット: violinplot
# - 棒グラフ: barplot
# - 散布図: jointplot
# - ペアプロット: pairplot(3つ以上の変数、散布図行列みたいなもの) ↓が例
iris = sns.load_dataset('iris')
iris.head(n=3)
```

```python
# カテゴリ: speciesの変数ごとに色分け
sns.pairplot(iris, hue='species')
```
